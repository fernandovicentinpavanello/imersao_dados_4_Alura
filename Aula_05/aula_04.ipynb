```python
# Vamos separar os dados da coluna Rua da nossa primeira base, pois temos o nome da rua e o número na mesma string. Vamos usar regex para separar esses dados.
dados_filtro["Rua"].str.extract(r'(^[\w ]+)')[:10]
```

```python
# Vamos criar uma nova coluna para visualizar melhor e poder trabalhar com a coluna. Usaremos novamente o regex para excluir qualquer vírgula, pontos, e vamos manter os endereços em letra minúscula. Assim temos o formato de endereço que queremos, igual a nossa base endereços_sp ( Base de CEPs ).
dados_filtro["apenas_rua"] = dados_filtro["Rua"].str.extract(r'(^[\w ]+)')
dados_filtro["apenas_rua"] = dados_filtro["apenas_rua"].str.lower().str.strip()
dados_filtro.head()
```
